```python
import numpy as np
import pandas as pd
np.random.seed(42)
n = 1000
total_tasks = np.random.randint(50, 300, n)
completed_tasks = np.random.randint(20, 250, n)
avg_task_delay = np.random.uniform(0, 10, n)  # days
budget_allocated = np.random.randint(50000, 500000, n)
budget_spent = budget_allocated * np.random.uniform(0.5, 1.2, n)
team_size = np.random.randint(3, 15, n)
team_experience = np.random.uniform(1, 10, n)  # years
sprint_velocity = np.random.uniform(10, 100, n)
past_delay_rate = np.random.uniform(0, 1, n)

# Derived features
completion_rate = completed_tasks / total_tasks
budget_burn_rate = budget_spent / budget_allocated
productivity_score = sprint_velocity / team_size

# Risk Logic (Label Engineering)
delay_score = (
    (1 - completion_rate) * 0.4 +
    budget_burn_rate * 0.3 +
    (avg_task_delay / 10) * 0.2 +
    (1 - productivity_score / productivity_score.max()) * 0.1 +
    past_delay_rate * 0.2
)

# Normalize score
delay_score = delay_score / delay_score.max()

# Binary target
delayed = (delay_score > 0.6).astype(int)

# Create dataframe
df = pd.DataFrame({
    "total_tasks": total_tasks,
    "completed_tasks": completed_tasks,
    "avg_task_delay": avg_task_delay,
    "budget_allocated": budget_allocated,
    "budget_spent": budget_spent,
    "team_size": team_size,
    "team_experience": team_experience,
    "sprint_velocity": sprint_velocity,
    "past_delay_rate": past_delay_rate,
    "completion_rate": completion_rate,
    "budget_burn_rate": budget_burn_rate,
    "productivity_score": productivity_score,
    "delayed": delayed
})

df.to_csv("project_data.csv", index=False)

print("Dataset generated successfully!")
print(df.head())
```
